# Problem 1: Social Media – Post Engagement Analyzer

Problem Statement:
Create a function to analyze engagement on a social media post.

Rules
Calculate total likes
If Total Likes ≥ 1000 → Viral Post
Otherwise → Normal Engagement

Real-Time Use
Social media analytics platforms
Marketing dashboards

Hint
Use function
Use loop for calculations

Expected Output Format
Total Likes: 1050
Post Status: Viral Post



```python
def analyze_engagement(likes):    #likes is a list of integers representing the number of likes on a post
    total_likes = 0
    for i in likes:               #iterating through the list of likes and adding them to total_likes
        total_likes+=i
    if total_likes >= 1000:        #if total_likes is greater than or equal to 1000, it is considered a viral post
        print(f"Total Likes: {total_likes}")
        print("Post Status: Viral Post")
        
    else:
        print(f"Total Likes: {total_likes}")
        print("Post Status: Normal Engagement")
        

likes = [100, 200, 300, 400,50]
analyze_engagement(likes)
```

# Problem 3: Agriculture – Rainfall Adequacy Checker

Problem Statement 
Create a function that analyzes seasonal rainfall data.

Rules Calculate average rainfall

Average ≥ required level → Adequate Rainfall

Otherwise → Inadequate Rainfall

Real-Time Use

Climate monitoring systems

Irrigation planning tools

Hint

Use function

Use loop and math operations

Expected Output Format

Average Rainfall: 72

Rainfall Status: Adequate Rainfall

```python
def seasonal_rainfall(rainfall_data):      #rainfall_data is a list of integers representing the rainfall measurements for a season
  required_level = 90
  total = 0
  for i in rainfall_data:                  #iterating through the list of rainfall data and adding them to total
    total+=i
  avg_rainfall= total/len(rainfall_data)
  print(f"Average Rainfall: {avg_rainfall}")
  if avg_rainfall >=required_level:              #if the average rainfall is greater than or equal to the required level, it is considered adequate rainfall
      print("Rainfall Status: Adequate Rainfall")
  else:
      print("Rainfall Status: Inadequate Rainfall")

rainfall_data = [95,96,80,91,90]
seasonal_rainfall(rainfall_data)

```

# Problem 4: Social Media – Duplicate Account Detection

Problem Statement

Create a function to detect duplicate usernames in a system.

Rules

If duplicates exist → Flag accounts

Else → Accounts are unique

Real-Time Use

Platform moderation

Fraud detection systems

Hint

Use function

Use set data structure

Expected Output Format

Duplicate Accounts Found: Yes

```python
def Duplicate_Account_Detection(usernames):                #usernames is a list of strings representing the usernames of accounts
  unique_usernames = set()                      #creating an empty set to store unique usernames
  Flag_accounts = False                #flag to indicate if duplicate accounts are found
  for username in usernames:               #iterating through the list of usernames and checking if they are already in the unique_usernames set, if they are, it means a duplicate account is found and the flag is set to True, otherwise, the username is added to the unique_usernames set
    if username in unique_usernames:
      Flag_accounts = True
    else:
      unique_usernames.add(username)
  if Flag_accounts:
    print("Duplicate Accounts Found: Yes")
  else:
    print("Accounts are unique")

usernames = ["user1","user2","user3","user1","user3"]
Duplicate_Account_Detection(usernames)
            
```

Problem 6: Agriculture – Premium Crop Price Filter

Problem Statement
Create a function that filters crop prices above a premium threshold.

Rules
Price > ₹2000 → Premium Crop
Else → Regular Crop

Real-Time Use
Market analytics
Export pricing systems

Hint
Use function
Use list filtering logic

Expected Output Format
Premium Crops: [2500, 3200]


```python
def crop_price_filtering(crop_prices, threshold):        #crop_prices is a list of integers representing the prices of different crops, and threshold is an integer representing the minimum price for a crop to be considered premium
    premium_crop = []
    for crop_price in crop_prices:                       #iterating through the list of crop prices and checking if they are greater than or equal to the threshold, if they are, they are added to the premium_crop list
        if crop_price >= threshold:
            premium_crop.append(crop_price)
    return premium_crop
threshold = 800
crop_prices = [1000,2000,3900,500,350]
result = crop_price_filtering(crop_prices, threshold)
print(f"Premium Crops:{result}")




        
    
```

# Problem 7: System Monitoring – Application Health Checker

Problem Statement
Create a function to evaluate application health based on error count.

Rules
Errors = 0 → Healthy
Errors ≤ 5 → Minor Issues
Errors > 5 → Critical Issues

Real-Time Use
System monitoring tools
IT operations dashboards

Hint
Use function
Use conditional statements

Expected Output Format

```python
def eval_health(error_count):               #error_count is an integer representing the number of errors detected in the system
    print(f"Error Count : {error_count}")
    if error_count == 0:                       #if there are no errors, the system is considered healthy
          print("System Status: Healthy")
    elif error_count <= 5:                     #if there are 5 or fewer errors, the system is considered to have minor issues
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")

eval_health(7)
```

# Problem 9: E-Learning – Student Attendance Eligibility System

Problem Statement
Create a function to calculate attendance percentage and check exam eligibility.

Rules
Attendance ≥ 75% → Eligible
Attendance < 75% → Not Eligible

Real-Time Use
Learning management systems
Internship tracking platforms

Hint
Use function
Use list and loop

Expected Output Format
Attendance Percentage: 80.0
Exam Eligibility: Eligible


```python
def cal_percentage(present_days, total_days):
    total_present_days = 0
    
    # Using loop to calculate total present days
    for days in present_days:
        total_present_days += days
    
    percentage = (total_present_days / total_days) * 100
    
    print(f"Attendance Percentage: {percentage:.2f}")
    
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


present_days = [28, 18, 25, 15, 30]
total_days = 150

cal_percentage(present_days, total_days)
```

# Problem 10: Smart Electricity Bill Analyzer

Problem Statement
You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.

Billing Rules
First 100 units → ₹3 per unit
Next 100 units (101–200) → ₹5 per unit
Above 200 units → ₹7 per unit
Usage Classification
Total bill < ₹500 → Low Usage
₹500 – ₹1500 → Moderate Usage
₹1500 → High Usage

Real-Time Use
Smart meter systems
Electricity board billing software
Energy consumption analytics

Hint
Use function to calculate bill
Use conditional logic for slabs
Use return values (bill and usage status)

```python
def calculate_electricity_bill(units):
    bill = 0

    # Slab calculation
    if units <= 100:
        bill = units * 3

    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)

    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    return bill, usage_status


# Example
total_bill, status = calculate_electricity_bill(250)

print(f"Total Bill: ₹{total_bill}")
print(f"Usage Level: {status}")
```
