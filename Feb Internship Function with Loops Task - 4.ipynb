### Problem 1: Mobile Recharge Validation System

Problem Statement
Create a function to validate a mobile recharge request.

Rules:
Recharge amount must be ≥ ₹50
Valid plans: 199, 299, 399, 599
User should be allowed to retry if invalid

Real-Time Use
Telecom recharge apps
Payment gateways

Hint
Use function for validation
Use while loop for retry


```python
## Function to validate recharge amount
def validation(recharge_amount):   
    valid_plans = [199,299,399,599]

    ## Rule 1: Minimum recharge check
    if recharge_amount < 50:
        print("Minumum recharge is 50")
        return False
    
    # Rule 2: Check if amount is in valid plans
    elif recharge_amount not in valid_plans:
        print("Invalid plan")
        return False
    
       # If all conditions pass
    else:
        return True
    
 #Retry system using while loop   
while True:
       recharge_amount = int(input("Enter Amount: "))

       # Call validation function
       if validation(recharge_amount):
            print("Recharge successful")
            break
       else:
            print("Try Again")
     
        
```

### Problem 2: Inventory Reorder Alert System
Problem Statement
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15


Print Stock OK otherwise


Real-Time Use
Warehouse systems


Retail management


Hint
Use function


Use for loop on dictionary

```python
# Function to check stock levels
def check_stock(products):
    
    # Loop through each product and its quantity in the dictionary
    for product, quantity in products.items():
        
        # If stock is less than 15 → Reorder Alert
        if quantity < 15:
            print(f"{product}: Reorder Alert (Stock = {quantity})")
        
        # Otherwise → Stock OK
        else:
            print(f"{product}: Stock OK (Stock = {quantity})")


# Dictionary containing product names and their stock quantities
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 30,
    "Chair": 8
}

# Calling the function
check_stock(inventory)
```

###  Problem 3: Student Result Processing System
Problem Statement
Create a function that:
Accepts a list of student marks


Calculates average marks


Prints Pass if average ≥ 50 else Fail


Real-Time Use
Online exam systems


Evaluation portals


Hint
Use loop to calculate total


Return result from function


```python
# Function to calculate average and return result
def evaluate_student(marks):

    total = 0   # Variable to store total marks

    # Loop to calculate total marks
    for mark in marks:
        total += mark

    # Calculate average
    average = total / len(marks)

    # Check pass or fail condition
    if average >= 50:
        return f"Average = {average:.2f} → Pass"
    else:
        return f"Average = {average:.2f} → Fail"


# Example list of student marks
student_marks = [60, 45, 70, 55]

# Calling the function and printing result
result = evaluate_student(student_marks)
print(result)
```

### Problem 4: Cab Fare Estimator with Retry Option
Problem Statement
Create a cab fare calculator:
Base fare = ₹50


₹12 per km


Peak hour → 25% extra


Allow retry using while loop


Real-Time Use
Ride-hailing apps


Transport systems


Hint
Function for fare calculation

While loop for retry

```python
# Function to calculate cab fare
def calculate_fare(distance, peak_hour):

    base_fare = 50
    per_km_rate = 12

    # Calculate basic fare
    fare = base_fare + (distance * per_km_rate)

    # Add 25% extra if peak hour
    if peak_hour.lower() == "yes":
        fare = fare + (fare * 0.25)

    return fare


# Retry system using while loop
while True:

    distance = float(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak_hour)

    print(f"Total Cab Fare: ₹{total_fare:.2f}")

    retry = input("Do you want to calculate again? (yes/no): ")

    if retry.lower() != "yes":
        print("Thank you!")
        break
```

### Problem 5: Employee Attendance Eligibility Checker
Problem Statement
Create a function that:
Accepts attendance list ("P", "A")


Calculates attendance percentage


Returns Eligible if ≥ 75% else Not Eligible


Real-Time Use
HR payroll systems


Internship tracking


Hint
Use loop to count present days


Use return statement


```python
# Function to check attendance eligibility
def check_attendance(attendance_list):

    present_count = 0   # To count number of "P"

    # Loop to count present days
    for day in attendance_list:
        if day == "P":
            present_count += 1

    # Calculate attendance percentage
    total_days = len(attendance_list)
    percentage = (present_count / total_days) * 100

    # Check eligibility
    if percentage >= 75:
        return f"Attendance = {percentage:.2f}% → Eligible"
    else:
        return f"Attendance = {percentage:.2f}% → Not Eligible"


# Example attendance list
attendance = ["P", "A", "P", "P", "P", "A", "P", "P"]

# Calling function
result = check_attendance(attendance)
print(result)
```

### Problem 6: Password Strength Checker
Problem Statement
Create a function to check password strength.
Rules:
Minimum length = 8


Must contain at least one digit


Must contain at least one special character (@#$)


Real-Time Use
Login systems


Security validation


Hint
Use loop over string

Use conditions inside function

```python
# Function to check password strength
def check_password(password):

    # Rule 1: Minimum length check
    if len(password) < 8:
        return "Weak Password (Minimum length is 8)"

    has_digit = False
    has_special = False

    # Loop through each character
    for char in password:

        # Check if character is a digit
        if char.isdigit():
            has_digit = True

        # Check if character is special (@#$)
        if char in "@#$":
            has_special = True

    # Final validation
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character)"
        

# Example usage
password_input = input("Enter password: ")
result = check_password(password_input)
print(result)
```
